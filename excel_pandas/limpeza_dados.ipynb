```python
import pandas as pd

# ler excel
df = pd.read_excel("base-bruta.xlsx")

# dividir a coluna A pelas vírgulas
df = df.iloc[:,0].str.split(",", expand=True)

# definir colunas
df.columns = [
    "ID_Produto",
    "Nome_Produto",
    "Categoria",
    "Data_Registro",
    "Estoque_Atual",
    "Estoque_Minimo",
    "Custo_Unitario",
    "Venda_Media_Diaria",
    "Lead_Time_Fornecedor",
    "Status_Entrega"
]

# substituir NULL por valor nulo
df = df.replace("NULL", pd.NA)

# converter tipos
df["Data_Registro"] = pd.to_datetime(df["Data_Registro"], errors="coerce")

# salvar arquivo tratado
df.to_excel("dados-tratados.xlsx", index=False)

df.head()
```
